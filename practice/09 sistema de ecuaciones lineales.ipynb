```python
x = np.arange(-6,6)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x+1
```

```python
plt.figure()
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")

plt.show()
```

```python
v1 = np.array([2,5])
v2 = np.array([3,2])
```

```python
def graficarVectores(vecs,cols,alpha=1):

    plt.figure()
    plt.axvline(x=0,color="grey",zorder=0)
    plt.axhline(y=0,color="grey",zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                    angles='xy', scale_units='xy', scale=1, color=cols[i], alpha=alpha
                   )
```

```python
%run "./funcionesAuxiliares/graficarVectores.ipynb"
```

```python
graficarVectores([v1,v2],["orange","blue"])
plt.xlim(-1,8)
plt.ylim(-1,8)
```

```python
v1v2 = 2*v1+1*v2
print(v1v2)
```

```python
graficarVectores([v1,v2,v1v2],["orange","blue","red"])
plt.xlim(-1,8)
plt.ylim(-1,12)
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                    marker='|',
                    color='orange')
plt.xlim(-100,100)
plt.ylim(-100,100)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```
