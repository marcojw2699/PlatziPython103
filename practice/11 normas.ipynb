```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
v1 = np.array([2,7])
v2 = np.array([3,5])
print(v1)
print(v2)
```

```python
v1v2 = v1 + v2
print(v1v2)
```

```python
print(np.linalg.norm(v1))
print(np.linalg.norm(v2))
print(np.linalg.norm(v1)+np.linalg.norm(v2))
print(np.linalg.norm(v1v2))
```

```python
v1 = np.array([0,0,2,7])
v2 = np.array([0,0,3,5])

v1_aux = np.array([v1[2],v1[3],v2[2],v2[3]])
v1v2 = v1 + v2 

print(v1_aux)
print(v1v2)
```

```python
plt.quiver([v1[0],v1_aux[0],v1v2[0]],
           [v1[1],v1_aux[1],v1v2[1]],
           [v1[2],v1_aux[2],v1v2[2]],
           [v1[3],v1_aux[3],v1v2[3]],
           angles='xy', scale_units='xy', scale = 1,
           color=sns.color_palette()
           )

plt.xlim(-0.5,6)
plt.ylim(-0.5,15)
```

```python
vector = np.array([1,2,0,5,6,0])

print(np.linalg.norm(vector,ord=0))
```

```python
vector = np.array([1,-1,1,-1,1])
print(np.linalg.norm(vector,ord=1))
```

```python
v1 = np.array([0,0,0,3])
v2 = np.array([0,0,3,3])
```

```python
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver([v1[0],v2[0]],
           [v1[1],v2[1]],
           [v1[2],v2[2]],
           [v1[3],v2[3]],
           angles='xy', scale_units='xy', scale=1, color = sns.color_palette()
           )

plt.show()
```
